### 2) Read Image

```python
img = cv2.imread('car1.jpeg')
```

```python
# cv2.imshow('Car image', img) 
# => DisabledFunctionError: cv2.imshow() is disabled in Colab, because it causes Jupyter sessions
```

```python
# solution 1: Matplotlib
import matplotlib.pyplot as plt
plt.imshow(img)
```

원래 빨간색 자동차인데 파란색으로 표시된다. 그 이유는 OpenCV의 컬러 표현 방법이 기본적인 RGB가 아니라 BGR이기 때문이다. 이는 다음과 같은 방법으로 해결할 수 있다.

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

이뿐만 아니라 다양한 색 변환이 있다.

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img)
```

```python
# Solution 2: cv2_imshow => Colab recommended
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

자동으로 RGB 변환까지 해주는 것을 확인할 수 있다.

```python
cv2_imshow(rgb_img)
```

오히려 rgb_img를 출력하니 한번 더 바뀌어 BGR 이미지가 출력되는 것을 확인할 수 있다.

```python
cv2_imshow(gray_img)
```

```python
# Solution 3: PIL
import PIL.Image as Image
Image.fromarray(img)
```
