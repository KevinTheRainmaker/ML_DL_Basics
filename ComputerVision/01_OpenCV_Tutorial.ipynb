OpenCV는 컴퓨터비전을 목적으로 하는 오픈소스 라이브러리이다. IPP를 지원하며, 기존 C++로 구현되었으나 OpenCV-python을 통해 Python 포팅이 되어 있다.

### 2) Read Image

```python
img = cv2.imread('car1.jpg')
img
```

```python
# cv2.imshow('Car image', img) 
# => DisabledFunctionError: cv2.imshow() is disabled in Colab, because it causes Jupyter sessions
```

```python
# solution 1: Matplotlib
import matplotlib.pyplot as plt
plt.imshow(img)
```

원래 빨간색 자동차인데 파란색으로 표시된다. 그 이유는 OpenCV의 컬러 표현 방법이 기본적인 RGB가 아니라 BGR이기 때문이다. 이는 다음과 같은 방법으로 해결할 수 있다.

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

이뿐만 아니라 다양한 색 변환이 있다.

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
```

```python
# Solution 2: cv2_imshow => Colab recommended
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

자동으로 RGB 변환까지 해주는 것을 확인할 수 있다.

```python
cv2_imshow(rgb_img)
```

오히려 rgb_img를 출력하니 한번 더 바뀌어 BGR 이미지가 출력되는 것을 확인할 수 있다.

```python
cv2_imshow(gray_img)
```

```python
# Solution 3: PIL
import PIL.Image as Image
Image.fromarray(img)
```

### +) 추가

보기 쉽게 한 필드에 이미지를 출력해보자.

```python
plt.subplot(1,3,1) # (nrows, ncols, index)
plt.imshow(img)
plt.axis('off')
plt.title('Image - original')

plt.subplot(1,3,2)
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.axis('off')
plt.title('Image - RGB')

plt.subplot(1,3,3)
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
plt.axis('off')
plt.title('Image - Gray')
```
